```python
def register_employee():
    print("\n===== Employee Registration =====")

    emp_id = input("Enter Employee ID: ")
    name = input("Enter Name: ")
    age = int(input("Enter Age: "))
    department = input("Enter Department: ")
    basic_salary = float(input("Enter Basic Salary: "))

    if age < 18:
        print("❌ Error: Age must be 18 or above")
        return None

    if basic_salary <= 0:
        print("❌ Error: Salary must be greater than 0")
        return None

    print("✅ Employee Registered Successfully")

    return emp_id, name, age, department, basic_salary


employee = register_employee()
if employee is None:
    exit()
```

```python
print("\n===== Attendance Management =====")
present_days = 0

for day in range(1, 6):
    status = input(f"Day {day} Attendance (P/A): ").upper()
    if status == "P":
        present_days += 1

print("Total Present Days:", present_days)

if present_days >= 4:
    attendance_status = "Regular"
elif present_days == 3:
    attendance_status = "Warning"
else:
    attendance_status = "Irregular"

print("Attendance Status:", attendance_status)
```

```python
def calculate_salary(basic_salary, present_days):
    hra = 0
    da = 0
    pf = basic_salary * 0.12

    if present_days >= 3:
        hra = basic_salary * 0.20
        da = basic_salary * 0.10

    net_salary = basic_salary + hra + da - pf
    return net_salary


net_salary = calculate_salary(employee[4], present_days)
print("Net Salary: ₹", net_salary)
```

```python
print("\n===== Performance Evaluation =====")
total_rating = 0

for i in range(1, 4):
    rating = int(input(f"Enter Rating {i} (1-5): "))
    total_rating += rating

average = total_rating / 3
print("Average Rating:", average)

if average >= 4:
    performance = "Excellent"
elif average >= 3:
    performance = "Good"
else:
    performance = "Needs Improvement"

print("Performance Status:", performance)
```
