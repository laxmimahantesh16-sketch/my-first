```python
# Daily Sales Aggregator

days = int(input("Enter number of days: "))

total_sales = 0

for i in range(1, days + 1):
    sales = float(input(f"Enter sales for day {i}: "))
    total_sales += sales

average_sales = total_sales / days

print("\n--- Sales Summary ---")
print("Total Sales:", total_sales)
print("Average Sales:", average_sales)
```

```python
# Inventory Reorder Checker

n = int(input("Enter number of products: "))

print("\n--- Inventory Status ---")

for i in range(1, n + 1):
    stock = int(input(f"Enter stock quantity for Product {i}: "))

    if stock < 20:
        print(f"Product {i}: Reorder Required")
    else:
        print(f"Product {i}: Stock Sufficient")
```

```python
# Attendance Summary Generator

n = int(input("Enter number of interns: "))

present_count = 0
absent_count = 0

for i in range(1, n + 1):
    attendance = input(f"Enter attendance for Intern {i} (P/A): ").upper()

    if attendance == 'P':
        present_count += 1
    elif attendance == 'A':
        absent_count += 1
    else:
        print("Invalid input! Please enter P or A.")

print("\n--- Attendance Summary ---")
print("Total Present:", present_count)
print("Total Absent:", absent_count)
```

```python
# Password Attempt Limiter

correct_password = "admin123"
attempts = 0
max_attempts = 3

while attempts < max_attempts:
    password = input("Enter password: ")

    if password == correct_password:
        print("Login Successful")
        break
    else:
        attempts += 1
        print("Incorrect password")

if attempts == max_attempts:
    print("Account Locked")
```
