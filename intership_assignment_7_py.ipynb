```python
import pandas as pd
import numpy as np

np.random.seed(42)

data = {
    "Emp_ID": range(101, 121),
    "Department": np.random.choice(
        ["HR", "IT", "Finance", "Sales"], 20),
    "Work_Hours": np.random.randint(6, 11, 20),
    "Attendance": np.random.randint(70, 101, 20),
    "Performance_Score": np.random.randint(50, 101, 20)
}

df = pd.DataFrame(data)
df.head()
# Overall averages
print("Average Work Hours:", df['Work_Hours'].mean())
print("Average Attendance:", df['Attendance'].mean())
print("Average Performance:", df['Performance_Score'].mean())

# Department-wise performance
dept_perf = df.groupby("Department")['Performance_Score'].mean()
print("\nDepartment-wise Performance:\n", dept_perf)

# Correlation
correlation = df[['Work_Hours','Attendance','Performance_Score']].corr()
print("\nCorrelation Matrix:\n", correlation)

# Top & Bottom employees
top_emp = df.nlargest(3, 'Performance_Score')
low_emp = df.nsmallest(3, 'Performance_Score')

top_emp, low_emp
```

```python
import pandas as pd
import numpy as np

np.random.seed(42)

data = {
    "Emp_ID": range(101, 121),
    "Department": np.random.choice(
        ["HR", "IT", "Finance", "Sales"], 20),
    "Work_Hours": np.random.randint(6, 11, 20),
    "Attendance": np.random.randint(70, 101, 20),
    "Performance_Score": np.random.randint(50, 101, 20)
}

df = pd.DataFrame(data)
df.head()
import matplotlib.pyplot as plt
import seaborn as sns

# Performance distribution
plt.hist(df['Performance_Score'], bins=10)
plt.title("Performance Score Distribution")
plt.xlabel("Score")
plt.ylabel("Employees")
plt.show()

# Department-wise performance
dept_perf.plot(kind='bar')
plt.title("Average Performance by Department")
plt.ylabel("Score")
plt.show()

# Work Hours vs Performance
sns.scatterplot(x='Work_Hours', y='Performance_Score', data=df)
plt.title("Work Hours vs Performance")
plt.show()

# Heatmap
sns.heatmap(correlation, annot=True)
plt.title("Correlation Heatmap")
plt.show()
```
