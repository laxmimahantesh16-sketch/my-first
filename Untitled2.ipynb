```python
# Accept Intern Details
name = input("Enter Intern Name: ")
age = int(input("Enter Age: "))
email = input("Enter Email ID: ")
contact = input("Enter Contact Number: ")
percentage = float(input("Enter Graduation Percentage: "))

# Validation
if age < 18:
    print("Rejected: Age must be 18 or above")

elif percentage < 60:
    print("Rejected: Graduation Percentage must be 60% or above")

elif len(contact) != 10 or not contact.isdigit():
    print("Rejected: Contact number must contain exactly 10 digits")

else:
    print("Intern Eligible for Internship")
```

```python
# Input Employee Details
emp_name = input("Enter Employee Name: ")
emp_id = input("Enter Employee ID: ")
basic_salary = float(input("Enter Basic Salary: "))

# Salary Calculations
hra = basic_salary * 0.20
da = basic_salary * 0.10
pf = basic_salary * 0.12

net_salary = basic_salary + hra + da - pf

# Display Salary Summary
print("\n------ Salary Summary ------")
print("Employee Name   :", emp_name)
print("Employee ID     :", emp_id)
print("Basic Salary    :", basic_salary)
print("HRA (20%)       :", hra)
print("DA (10%)        :", da)
print("PF (12%)        :", pf)
print("----------------------------")
print("Net Salary      :", net_salary)
```

```python
# Display available courses
print("Available Courses")
print("1. Python Programming")
print("2. Data Analytics")
print("3. AI & ML")

# User input
choice = int(input("Enter course number (1-3): "))
student = input("Are you a student? (yes/no): ")
early = input("Early registration? (yes/no): ")

# Course selection
if choice == 1:
    course_name = "Python Programming"
    fee = 5000
elif choice == 2:
    course_name = "Data Analytics"
    fee = 8000
elif choice == 3:
    course_name = "AI & ML"
    fee = 12000
else:
    print("Invalid course selection")
    exit()

# Discount calculation
discount = 0

if student.lower() == "yes":
    discount = discount + (fee * 0.10)

if early.lower() == "yes":
    discount = discount + (fee * 0.05)

final_fee = fee - discount

# Output
print("\n------ Course Fee Details ------")
print("Course Name        :", course_name)
print("Original Fee (₹)   :", fee)
print("Total Discount (₹) :", discount)
print("Final Payable (₹)  :", final_fee)
```

```python
# Input Employee Details
emp_name = input("Enter Employee Name: ")
salary = float(input("Enter Salary: "))
rating = int(input("Enter Performance Rating (1-5): "))

# Bonus Calculation
if rating == 5:
    bonus = salary * 0.20
elif rating == 4:
    bonus = salary * 0.15
elif rating == 3:
    bonus = salary * 0.10
else:
    bonus = 0

final_salary = salary + bonus

# Display Output
print("\n------ Bonus Evaluation Report ------")
print("Employee Name       :", emp_name)
print("Performance Rating  :", rating)
print("Bonus Amount (₹)    :", bonus)
print("Final Salary (₹)    :", final_salary)
```
