```python
def check_intern_eligibility(age, percentage):
    if age >= 18 and percentage >= 60:
        return "Eligible: Age is 18 or above and academic percentage meets the requirement."
    elif age < 18 and percentage < 60:
        return "Not Eligible: Age is below 18 and academic percentage is less than 60."
    elif age < 18:
        return "Not Eligible: Age is below 18."
    else:
        return "Not Eligible: Academic percentage is less than 60."
```
