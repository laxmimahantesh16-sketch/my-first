```python
import numpy as np

data = np.array([12, 15, 20, 20, 25, 30, 35, 40])
greater_than_20 = data[data > 20]
print("Values greater than 20:", greater_than_20)
```

```python
import numpy as np

data = np.array([12, 15, 20, 20, 25, 30, 35, 40])
count_less_20 = np.sum(data < 20)
print("Count of values less than 20:", count_less_20)
```

```python
import numpy as np

data = np.array([12, 15, 20, 20, 25, 30, 35, 40])
modified_data = data.copy()
modified_data[modified_data < 20] = 20
print("Modified data:", modified_data)
```
