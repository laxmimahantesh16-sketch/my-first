```python
# Inventory Stock Level Monitor

opening_stock = int(input("Enter opening stock: "))
items_sold = int(input("Enter items sold: "))

remaining_stock = opening_stock - items_sold

print("Remaining Stock:", remaining_stock)

if remaining_stock < 20:
    print("Reorder Required")
else:
    print("Stock Sufficient")
```

```python
# Delivery Distance Charge Calculator

distance = int(input("Enter delivery distance in km: "))

if distance <= 5:
    charge = 0
elif distance <= 15:
    charge = distance * 10
else:
    charge = distance * 20

print("Total Delivery Charge: ₹", charge)
```

```python
# Data Usage Alert System

total_data = float(input("Enter total data limit (GB): "))
data_used = float(input("Enter data used (GB): "))

usage_percentage = (data_used / total_data) * 100

if usage_percentage >= 90:
    print("High Usage Alert")
elif usage_percentage >= 70 and usage_percentage < 90:
    print("Warning")
else:
    print("Normal Usage")
```

```python
# Product Return Eligibility Checker

days = int(input("Enter days since delivery: "))
condition = input("Enter product condition (Good/Damaged): ")

if days <= 7 and condition == "Good":
    print("Return Accepted")
else:
    print("Return Rejected")
```
